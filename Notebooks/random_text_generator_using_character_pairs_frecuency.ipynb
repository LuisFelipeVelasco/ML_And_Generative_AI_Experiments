**LIBRARIES IMPORTATION**

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

**READ TEXT**

```python
with open('text_2.txt', 'r') as file:
    content = file.read()
    print(content)
```

**SAVE INDIVIDUAL CHARACTERS,ITS NUMBER OF OCCURRENCE AND ITS CUMULATIVE SUM OF OCCURRENCE**

```python
#convert all the text in lowercase
content=content.lower()
#convert the text into a list of characters
content_list=list(content)
#convert the list of characters into a numpy array
characters_list=np.array(content_list)
#Save each individual characters in the text and its number of occurrence
unique_chars,char_count=np.unique(characters_list,return_counts=True)
char_count_sum=np.cumsum(char_count)
print(unique_chars)
```

**CALCULATE THE NUMBER OF OCCURRENCES AND FRECUENCY OF EACH PAIR**

```python
#List to save each individual pair of characters in the text
character_pairs=[]
#List to save the number of occurrences of each pair of characters in the text
count_character_pairs=[]
#Catch all the 2 join characters of the text and count its occurrences
for i in range(0,len(content_list)-1):
  if((content_list[i]+content_list[i+1]) in character_pairs):
    count_character_pairs[character_pairs.index(content_list[i]+content_list[i+1])]+=1
  else:
    character_pairs.append(content_list[i]+content_list[i+1])
    count_character_pairs.append(1)
#Calculate the probability of each pair
frecuency_character_pairs=((np.array(count_character_pairs))*100/len(content_list)).tolist()
#Dictionary , wich keys are the individual pairs and its value the frecuency of that pair
result=dict(zip(character_pairs,frecuency_character_pairs))
```

**SHOW RESULTS**

```python
sorted_result = dict(sorted(result.items(), key=lambda item: item[1],reverse=True))
print(sorted_result)
```

**BUILD PAIR DISTRIBUTION PER INITIAL CHARACTER**

```python
# Dictionary that will store, for each initial character,
# the cumulative occurrence counts of all pairs that start with that character
pair_distribution_per_initial_char = {}

# Dictionary that will store, for each initial character,
# the list of pairs that start with that character
pairs_per_initial_char = {}

# Iterate through each unique character found in the text
for char in unique_chars:

    # Process the character only if it has not been added yet
    if char not in pairs_per_initial_char:

        # List to store the occurrence counts of pairs
        # that start with the current character
        ocurrence_pairs = []

        # List to store the pairs that start
        # with the current character
        pairs = []

        # Iterate through all discovered character pairs
        for index, pair in enumerate(character_pairs):

            # Check if the pair starts with the current character
            if pair[0] == char:

                # Store the occurrence count of the pair
                ocurrence_pairs.append(count_character_pairs[index])

                # Store the pair itself
                pairs.append(pair)

        cumulative_ocurrance_pairs=np.array(ocurrence_pairs).cumsum()

        # Save the occurrence counts for this character
        pair_distribution_per_initial_char[char] = cumulative_ocurrance_pairs

        # Save the corresponding pairs for this character
        pairs_per_initial_char[char] = pairs
```

**FUNCTION TO CHOOSE A RANDOM CHARACTER DEPENDING ON ITS FRECUENCY**

```python
def random_char(max_random_number,char_distribution):

  """
Parameters
----------
max_random_number : int
    Maximum  possible random number
char_distribution : numpy.ndarray
    Character distribution (%).

Returns
-------
unique_chars[index_random_char] : char
    Simulated character frequencies (%).
"""

  # Generate a random number within the frequency distribution range
  random_number = random.randint(0, max_random_number)

  # Find the index of the character corresponding to the random number
  index_random_char = np.searchsorted(char_count_sum, random_number)

  # Return the selected character
  return unique_chars[index_random_char]
```

**FUNCTION TO CHOOSE A RANDOM CHARACTER DEPENDING ON ITS FRECUENCY BEING JOINED TO A SPECIFIC CHARACTER**

```python
def random_second_char(pair_distribution,pairs,initial_char):

  """
Parameters
----------
pair_distribution : dict
  Cumulative frequency distribution of character pairs for each initial character.
pairs : dict
  Character pairs grouped by their initial character.
initial_char : char
  First character used to select the corresponding pair distribution.

Returns
-------
pairs[initial_char][index_random_second_char][1] : char
  Randomly selected second character based on the pair frequency distribution.
"""

  # Maximum possible random number for the selected initial character
  max_random_number = pair_distribution[initial_char][-1]

  # Generate a random number within the distribution range
  random_number = random.randint(0, max_random_number)

  # Find the index corresponding to the random number
  index_random_second_char = np.searchsorted(
      pair_distribution[initial_char],
      random_number
  )

  # Return the selected second character
  return pairs[initial_char][index_random_second_char][1]

```

**GENERATE RANDOM TEXT**

```python
# Generate the first character of the text according to the character distribution
text = random_char(len(content), char_count_sum)

# Continue generating characters until the desired text length is reached
while len(text) != len(content):

  # Get the last generated character
  first_char = text[-1]

  # Generate the next character based on the pair distribution
  second_char = random_second_char(
      pair_distribution_per_initial_char,
      pairs_per_initial_char,
      first_char
  )

  # Append the generated character to the text
  text += second_char

print(text)
```
