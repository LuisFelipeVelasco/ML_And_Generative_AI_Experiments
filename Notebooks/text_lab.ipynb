```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# **CALCULATING PROBABILITY OF CHARACTERS IN A TEXT**

**CHARGING TEXT**

```python
!wget https://raw.githubusercontent.com/LuisFelipeVelasco/Generative_AI_Experiments/refs/heads/main/text.txt
```

**READING TEXT**

```python
with open('text.txt', 'r') as file:
    content = file.read()
    print(content)
```

**CALCULATING THE PROBABILITY OF EACH CHARACTER**

```python
#convert all the text in lowercase
content=content.lower()
#Replace the spaces in content with_ and the endlines with \n
content_list=list(content)

#become each character into it´s ASCII code to work with numpy
characters_ASCII=np.array([ord(char) for char in content_list])
#Remove \n character
#characters_ASCII=np.delete(characters_ASCII,np.where(characters_ASCII==10))
#Calculate the probabilities:
unique_chars,char_count=np.unique(characters_ASCII,return_counts=True)
frecuency_chars=(char_count*100)/len(characters_ASCII)
```

**GRAPHING PROABIBLITIES**

```python
characters=[chr(code) for code in unique_chars]
for i,char in enumerate(characters):
    if(char=='\n'):
        characters[i]="\\n"
    if(char==' '):
        characters[i]="_"
plt.figure(figsize=(10, 20))
plt.bar(characters, frecuency_chars, color='skyblue')
plt.yticks(range(0, 16,1))
```

# **ANALAZYNG CHARACTER FRECUENCY USING RANDOM NUMBERS AND OCCURRENCE COUNT.**  

```python
number_random_numbers=len(characters_ASCII)
max_random_number=len(characters_ASCII)
#Generate a np array with random numbers
random_numbers=np.random.choice(max_random_number,number_random_numbers)
#Generate a np array with the cumulutive sum of char_count
char_count_sum=np.cumsum(char_count)
#Generate a np array with the index of the char that correspond to each random number
index_char=np.searchsorted(char_count_sum, random_numbers)
char_count_2=np.zeros(len(unique_chars))
unique_index_chars,index_char_frequencies=np.unique(index_char,return_counts=True)
# Map the simulated frequencies back to their absolute character index positions
char_count_2[unique_index_chars] = index_char_frequencies
frecuency_chars_2=(char_count_2*100)/len(random_numbers)
```

**GRAPHING PROBABILITIES**

```python
plt.figure(figsize=(10, 20))
plt.bar(characters, frecuency_chars_2, color='skyblue')
plt.yticks(range(0, 16,1))
```

# **CALCULATING PAIRS OF CHARACTERS FRECUENCY IN A TEXT**  


**CHARGING TEXT 2**

```python
!wget https://raw.githubusercontent.com/LuisFelipeVelasco/Generative_AI_Experiments/refs/heads/main/text_2.txt
```

**READING TEXT**

```python
with open('text_2.txt', 'r') as file:
    content_2 = file.read()
    print(content_2)
```

**CALCULATING THE PROBABILITY OF EACH PAIR**

```python
#convert all the text in lowercase
content_2=content_2.lower()
content_list=list(content_2)
#Change the spaces for_ and the endlines for \ to see them
for i,char in enumerate(content_list):
    if(char=='\n'):
        content_list[i]="\\"
    if(char==' '):
        content_list[i]="_"
character_pairs=[]
count_character_pairs=[]
#Catch all the 2 join characters of the text and count its frecuency
for i in range(0,len(content_list)-1):
  if((content_list[i]+content_list[i+1]) in character_pairs):
    count_character_pairs[character_pairs.index(content_list[i]+content_list[i+1])]+=1
  else:
    character_pairs.append(content_list[i]+content_list[i+1])
    count_character_pairs.append(1)
#Calculate the probability of each pair
frecuency_character_pairs=((np.array(count_character_pairs))*100/len(content_list)).tolist()
result=dict(zip(character_pairs,frecuency_character_pairs))
```

**SHOWING RESULTS**

```python
sorted_result = dict(sorted(result.items(), key=lambda item: item[1],reverse=True))
print(sorted_result)
```
